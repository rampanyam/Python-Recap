 ### A Simple Guessing Gaming Using Python 3.6
 
 ### The rules of the game are as follows: 
 
 
 
 
 The Challenge:

Write a program that picks a random integer from 1 to 100, and has players guess the number. The rules are:

1. If a player's guess is less than 1 or greater than 100, say "OUT OF BOUNDS"
2. On a player's first turn, if their guess is
 * within 10 of the number, return "WARM!"
 * further than 10 away from the number, return "COLD!"
3. On all subsequent turns, if a guess is 
 * closer to the number than the previous guess return "WARMER!"
 * farther from the number than the previous guess, return "COLDER!"
4. When the player's guess equals the number, tell them they've guessed correctly *and* how many guesses it took!

```python
print("Rules of the game")
print(" ")
print(" 1.  The main intent of this game is to guess the random number")

print(" ")
print("2. IF the Number is within 10 numbers of distance , it will appear as WARM or else COLD")

print("3. If your current input is better than the previous then the word WARMER will appear, or else COLDER ")

print(" All the best Mate")








import random
random_number = random.randint(1,100)



print("The random number is : ")
print(random_number)

guess_list = list()






while True:
    x = input("Enter Number")
    guess_list.append(int(x.strip()))
    difference = abs(int(x) - random_number)

        

    
    if x.strip().isdigit():
        
        if int(x) == random_number:
            print("Well Done, thats correct !!")
            print(" ")
            print("You took    " + str(len(guess_list)) + " guesses to Win")
            break
        
        elif abs(int(x) - random_number) <=10:
            print("WARM, Close Enough")
            print("Please Try Again ")
            print(" ")
            
            
        elif abs(int(x) - random_number) >10:
            print("COLD, oops not close")
            print("Please try again")
            print(" ")
       
      
            
        else:
            
            continue
        
        
        
        
        
        if len(guess_list) >1:
            if difference > abs(guess_list[-2] - random_number):
                print("You are going on the wrong path, COLDERRRRRR")
                print(" ")
                print("No worries, Try again Mate")
            else:
                print(" You are going on the right pathhh, WARMERRR")
                print(" ")
                print("No Dramas, you are getting there, Try Hard mate")
        else:
                pass
        
        
        
        
        
    else:
        print("Please re-type the guess is invalid")
        
```
